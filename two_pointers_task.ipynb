```python
def max_area(height):
    p1, p2 = 0, len(height) - 1
    result = 0

    while p1 < p2:
        width = p2 - p1
        square = min(height[p1], height[p2]) * width

        if square > result:
            result = square

    
        if height[p1] < height[p2]:
            p1 += 1
        else:
            p2 -= 1

    return result

print(max_area([1,8,6,2,5,4,8,3]))
```
